### **Q2. What is the difference between Population and Sample? Why is sampling important?**

**Answer:**

* **Population** â†’ the entire dataset (e.g., all users of Netflix).
* **Sample** â†’ a subset of the population (e.g., 1,000 randomly selected Netflix users).
* Sampling is important because analyzing an entire population is often impractical; the **Central Limit Theorem** ensures that sample means approximate normality.

### **Q3. Explain Mean, Median, and Mode with an example.**

**Answer:**
For data: `[2, 4, 4, 6, 8]`

* Mean = (2+4+4+6+8)/5 = 4.8
* Median = middle value = 4
* Mode = most frequent value = 4

ðŸ‘‰ Interviewers may ask: *When would you prefer Median over Mean?*
Answer: When data has **outliers or skewed distribution**.

### **Q4. What is Variance and Standard Deviation?**

**Answer:**

* **Variance** = average squared deviation from mean.
* **Standard deviation (SD)** = square root of variance (in same units as data).
  ðŸ‘‰ High variance/SD = data is spread out, low variance/SD = data is tightly clustered.

### **Q7. What are Type I and Type II Errors?**

**Answer:**

* **Type I Error (False Positive):** Rejecting a true null hypothesis.
* **Type II Error (False Negative):** Failing to reject a false null hypothesis.
  ðŸ‘‰ Example: In medical tests, a **Type I error** = diagnosing disease when patient is healthy; **Type II error** = missing a real disease.

### **Q8. What is Correlation vs Covariance?**

**Answer:**

* **Covariance** â†’ measures how two variables vary together.
* **Correlation** â†’ normalized version of covariance, always between **-1 and +1**.
  ðŸ‘‰ Correlation is dimensionless and easier to interpret.

### **Q9. What is the Central Limit Theorem (CLT)? Why is it important?**

**Answer:**

* CLT states that the **sampling distribution of the sample mean** tends to be **normal**, regardless of population distribution, as sample size increases (n > 30 is a rule of thumb).
* Importance: Enables us to apply **hypothesis testing, confidence intervals** in ML/AI.

**Answer:**

* Examples:

  * Continuous: Normal, Exponential, Uniform.
    ðŸ‘‰ Normal distribution is crucial in ML because many algorithms assume normality.
