### **Q1. What is a list in Python?**

**Answer:**
A list is a **mutable, ordered collection** of items. It can store elements of **different data types**, including other lists.

* Indexed (supports positive and negative indexing).
* Preserves insertion order.
* Allows duplicates.

ğŸ“Œ Example:

```python
lst = [10, "AI", 3.14, [1, 2]]
print(lst[0])   # 10
print(lst[-1])  # [1, 2]
```

### **Q2. How do you create a list in Python?**

**Answer:**
Lists can be created in multiple ways:

```python
lst1 = [1, 2, 3]                # Literal
lst2 = list((4, 5, 6))          # Using list() constructor
lst3 = [0] * 5                  # Repetition â†’ [0,0,0,0,0]
lst4 = [i*i for i in range(5)]  # List comprehension â†’ [0,1,4,9,16]
```

### **Q4. How do you modify a list?**

**Answer:**
Lists are mutable, so elements can be changed, added, or removed:

```python
lst = [1, 2, 3]
lst[1] = 20            # Modify element â†’ [1,20,3]
lst.append(4)          # Add at end â†’ [1,20,3,4]
lst.insert(1, 15)      # Insert at index â†’ [1,15,20,3,4]
lst.extend([5,6])      # Add multiple â†’ [1,15,20,3,4,5,6]
lst.remove(20)         # Remove by value â†’ [1,15,3,4,5,6]
popped = lst.pop(2)    # Remove by index â†’ [1,15,4,5,6], popped=3
del lst[0]             # Delete element at index â†’ [15,4,5,6]
lst.clear()            # Empty list â†’ []
```

### **Q6. Whatâ€™s the difference between `append()` and `extend()`?**

**Answer:**

* `append(x)` â†’ adds the whole object as a **single element**.
* `extend(iterable)` â†’ iterates and adds each element individually.

ğŸ“Œ Example:

```python
lst = [1,2]
lst.append([3,4])   # [1,2,[3,4]]
lst = [1,2]
lst.extend([3,4])   # [1,2,3,4]
```

### **Q7. How does slicing work in lists?**

**Answer:**
Slicing uses `[start:end:step]` format.

```python
lst = [0,1,2,3,4,5]
print(lst[1:4])     # [1,2,3]
print(lst[:3])      # [0,1,2]
print(lst[::2])     # [0,2,4]
print(lst[::-1])    # Reverse â†’ [5,4,3,2,1,0]
```

### **Q8. What are shallow vs deep copies in lists?**

**Answer:**

* **Shallow copy** â†’ copies only references (nested lists still point to same object).
* **Deep copy** â†’ creates independent copies of all nested objects.

ğŸ“Œ Example:

```python
import copy

lst1 = [[1,2],[3,4]]
shallow = lst1.copy()
deep = copy.deepcopy(lst1)

lst1[0][0] = 99
print(shallow)  # [[99,2],[3,4]] (changed)
print(deep)     # [[1,2],[3,4]]  (unchanged)
```

### **Q9. Why is multiplying lists sometimes dangerous?**

**Answer:**
Because it replicates **references** for mutable objects.

```python
lst = [[0]] * 3
lst[0][0] = 1
print(lst)   # [[1],[1],[1]]  (all changed!)
```

âœ… Fix:

```python
lst = [[0] for _ in range(3)]
```

### **Q10. How do lists differ from tuples, sets, and arrays?**

**Answer:**

* **List** â†’ ordered, mutable, allows duplicates.
* **Tuple** â†’ ordered, immutable.
* **Set** â†’ unordered, unique elements.
* **Array (from `array` module)** â†’ more memory-efficient, type-restricted.

### **Q11. Whatâ€™s the time complexity of list operations?**

**Answer:**

| **Operation**        | **Time Complexity** |
| -------------------- | ------------------- |
| Indexing             | `O(1)`              |
| Append               | Amortized `O(1)`    |

### **Q12. How do you flatten a nested list?**

**Answer:**
Using list comprehension or recursion:

```python

flat = [x for sub in nested for x in sub]
print(flat)  # [1,2,3,4,5,6]
```

### **1. Two Sum (LeetCode #1)**

**Problem:** Given an array of numbers, return indices of the two numbers that add up to a target.

```python
def twoSum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        diff = target - num
        if diff in seen:
            return [seen[diff], i]
        seen[num] = i

print(twoSum([2,7,11,15], 9))  # [0,1]
```

### **2. Remove Duplicates from Sorted Array (LeetCode #26)**

**Problem:** Modify the array in place so each element appears once.

```python
def removeDuplicates(nums):
    if not nums: return 0
    i = 0
    for j in range(1, len(nums)):
        if nums[j] != nums[i]:
            i += 1
            nums[i] = nums[j]
    return i+1

print(removeDuplicates([1,1,2,2,3]))  # 3 (array becomes [1,2,3])
```

### **3. Move Zeroes (LeetCode #283)**

**Problem:** Move all zeroes to the end while maintaining relative order.

```python
def moveZeroes(nums):
    nonzero = 0
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[nonzero], nums[i] = nums[i], nums[nonzero]
            nonzero += 1
    return nums

print(moveZeroes([0,1,0,3,12]))  # [1,3,12,0,0]
```

### **4. Product of Array Except Self (LeetCode #238)**

**Problem:** Return array `output[i]` where each is product of all elements except itself.

```python
def productExceptSelf(nums):
    n = len(nums)
    res = [1] * n
    left = 1
    for i in range(n):
        res[i] = left
        left *= nums[i]
    right = 1
    for i in range(n-1, -1, -1):
        res[i] *= right
        right *= nums[i]
    return res

print(productExceptSelf([1,2,3,4]))  # [24,12,8,6]
```

### **6. 3Sum (LeetCode #15)**

**Problem:** Find all unique triplets such that their sum is zero.

```python
def threeSum(nums):
    nums.sort()
    res = []
    for i in range(len(nums)):
        if i > 0 and nums[i] == nums[i-1]:
            continue
        l, r = i+1, len(nums)-1
        while l < r:
            total = nums[i] + nums[l] + nums[r]
            if total == 0:
                res.append([nums[i], nums[l], nums[r]])
                l += 1; r -= 1
                while l < r and nums[l] == nums[l-1]:
                    l += 1
            elif total < 0:
                l += 1
            else:
                r -= 1
    return res

print(threeSum([-1,0,1,2,-1,-4]))  
# [[-1,-1,2],[-1,0,1]]
```

### **7. Trapping Rain Water (LeetCode #42)**

**Problem:** Compute how much water can be trapped after raining.

```python
def trap(height):
    if not height: return 0
    l, r = 0, len(height)-1
    left_max, right_max = height[l], height[r]
    water = 0
    while l < r:
        if left_max < right_max:
            l += 1
            left_max = max(left_max, height[l])
            water += left_max - height[l]
        else:
            r -= 1
            right_max = max(right_max, height[r])
            water += right_max - height[r]
    return water

print(trap([0,1,0,2,1,0,1,3,2,1,2,1]))  # 6
```

### **8. Merge Intervals (LeetCode #56)**

**Problem:** Merge overlapping intervals.

```python
def merge(intervals):
    intervals.sort(key=lambda x: x[0])
    merged = []
    for interval in intervals:
        if not merged or merged[-1][1] < interval[0]:
            merged.append(interval)
        else:
            merged[-1][1] = max(merged[-1][1], interval[1])
    return merged

print(merge([[1,3],[2,6],[8,10],[15,18]]))  # [[1,6],[8,10],[15,18]]
```

### **9. Subarray Sum Equals K (LeetCode #560)**

**Problem:** Find total number of continuous subarrays with sum equal to `k`.

```python
from collections import defaultdict

def subarraySum(nums, k):
    count = 0
    prefix_sum = 0
    freq = defaultdict(int)
    freq[0] = 1

    for num in nums:
        prefix_sum += num
        count += freq[prefix_sum - k]
        freq[prefix_sum] += 1

    return count

print(subarraySum([1,1,1], 2))  # 2
```
