## üîπ **Numeric Types in Python**

Python has **three main numeric types**:

1. **`int`** ‚Äì integers (arbitrary precision).
2. **`float`** ‚Äì floating-point numbers (double precision, IEEE 754).
3. **`complex`** ‚Äì numbers with real and imaginary parts (`a+bj`).

üìå Example:

```python
a = 10        # int
b = 3.14      # float
c = 2 + 3j    # complex
```

‚úÖ All numeric types are **immutable** and **hashable**.

## üîπ **Basic Numeric Operations**

| **Operation**  | **Example**         | **Output**           |
| -------------- | ------------------- | -------------------- |
| Addition       | `5 + 3`             | `8`                  |
| Subtraction    | `5 - 3`             | `2`                  |
| Multiplication | `5 * 3`             | `15`                 |
| Division       | `5 / 2`             | `2.5` (always float) |
| Floor Division | `5 // 2`            | `2`                  |

| Absolute value | `abs(-7)`           | `7`                  |

| divmod         | `divmod(5, 2)`      | `(2, 1)`             |

## üîπ **Conversions**

```python
int(3.9)     # 3
float(5)     # 5.0
complex(2)   # (2+0j)
```

‚ö° Trick: `int("010")` ‚Üí `10`, but `int("010", 8)` ‚Üí `8` (octal).

## üîπ **Math Functions (`math` module)**

```python
import math

print(math.sqrt(16))    # 4.0
print(math.factorial(5)) # 120
print(math.gcd(12, 18)) # 6
print(math.floor(3.7))  # 3
print(math.ceil(3.7))   # 4
print(math.pi)          # 3.141592...
```

---

## üîπ **Complex Numbers (`cmath` module)**

```python
import cmath

z = 2 + 3j
print(z.real)       # 2.0
print(z.imag)       # 3.0
print(abs(z))       # 3.6055 (magnitude)
print(cmath.phase(z))  # angle in radians
```

## üîπ **Tricky Interview Questions**

### **Q1. Why `5 / 2 = 2.5` but `5 // 2 = 2`?**

* `/` ‚Üí true division (always float).
* `//` ‚Üí floor division (rounds down towards negative infinity).

üìå Example:

```python
print(-5 // 2)  # -3 (not -2, because of floor behavior)
```

---

### **Q2. Why `0.1 + 0.2 != 0.3` in Python?**

Because floats use **binary floating-point representation** (IEEE 754). Some decimals cannot be represented exactly.

```python
print(0.1 + 0.2)       # 0.30000000000000004
```

‚úÖ Fix:

```python
from decimal import Decimal
print(Decimal("0.1") + Decimal("0.2"))  # 0.3
```

---

### **Q3. Difference between `is` and `==` with numbers?**

```python
a = 256
b = 256
print(a is b)   # True (small int caching)
x = 257
y = 257
print(x is y)   # False (outside cache range)
```

‚úÖ `==` ‚Üí value equality.
‚úÖ `is` ‚Üí memory identity.

---

### **Q4. Why does `True + True == 2`?**

Because `bool` is a subclass of `int`.

```python
print(True + True)  # 2
print(isinstance(True, int))  # True
```

---

### **Q5. How to handle very large numbers in Python?**

* Python‚Äôs `int` is **arbitrary precision** (no overflow).

```python
print(10**100)  # Very large int works
```

* But **float** has limits (`1.8e308` max). Beyond that ‚Üí `inf`.

---

### **Q6. Difference between `math.isnan()` and `math.isinf()`?**

```python
import math
print(math.isnan(float("nan")))  # True
print(math.isinf(float("inf")))  # True
```

---

### **Q7. Explain `Decimal` vs `Fraction`.**

* **`decimal.Decimal`** ‚Üí exact decimal arithmetic (financial apps).
* **`fractions.Fraction`** ‚Üí rational numbers.

```python
from decimal import Decimal
from fractions import Fraction

print(Decimal("0.1") + Decimal("0.2"))   # 0.3
print(Fraction(1, 3) + Fraction(1, 6))   # 1/2
```
