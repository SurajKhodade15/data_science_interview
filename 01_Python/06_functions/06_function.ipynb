**Answer (Interview Style):**
A function is a reusable block of code that performs a specific task. Functions in Python are defined using the `def` keyword and can take inputs (parameters) and return outputs.

## ðŸ”¹ **Q3. Function Arguments in Python**

Python supports **multiple argument types**:

| Type                              | Example                      | Usage                               |
| --------------------------------- | ---------------------------- | ----------------------------------- |
| **Positional**                    | `def add(a, b)` â†’ `add(2,3)` | Must match order                    |
| **Keyword**                       | `add(b=3, a=2)`              | Order doesnâ€™t matter                |
| **Default**                       | `def greet(name="Guest")`    | Uses default if not passed          |
| **Variable Positional (\*args)**  | `def fun(*args)`             | Collects multiple args into a tuple |

## ðŸ”¹ **Q4. What is Scope in Python? (LEGB Rule)**

Python resolves variables using **LEGB**:

* **L** â†’ Local (inside function)
* **E** â†’ Enclosing (outer functions)
* **G** â†’ Global (module-level)
* **B** â†’ Built-in (`len`, `print`)

```python
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print(x)
    inner()

outer()  # local
```

## ðŸ”¹ **Q5. Lambda Functions**

Anonymous, single-expression functions:

```python
square = lambda x: x*x
print(square(5))  # 25
```

Common in `map()`, `filter()`, `sorted()`:

```python
nums = [1,2,3,4]
print(list(map(lambda x: x**2, nums)))  # [1,4,9,16]
```

## ðŸ”¹ **Q6. Recursion Example**

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(5))  # 120
```

## ðŸ”¹ **Q7. Higher-Order Functions**

Functions that take other functions as arguments or return them.

```python
def apply_twice(f, x):
    return f(f(x))

print(apply_twice(lambda x: x+1, 5))  # 7
```

## ðŸ”¹ **Q8. Decorators**

Decorators are functions that wrap another function to extend its behavior.

```python
def log(func):
    def wrapper(*args, **kwargs):
        print("Calling:", func.__name__)
        return func(*args, **kwargs)
    return wrapper

@log
def greet(name):
    return f"Hi {name}"

print(greet("Suraj"))
```

# **Function Annotations in Python**

---

## ðŸ”¹ **Q1. What are function annotations?**

**Answer (Interview Style):**
Function annotations in Python let you attach **metadata** (usually type hints) to function parameters and return values.

* They donâ€™t affect execution.
* They are **hints only**, used by tools like **mypy, PyCharm, IDEs, linters, and documentation generators**.

---

## ðŸ”¹ **Q2. Basic Syntax**

```python
def greet(name: str, age: int) -> str:
    return f"Hello {name}, you are {age} years old."
```

* `name: str` â†’ parameter `name` is expected to be a string.
* `age: int` â†’ parameter `age` is expected to be an integer.
* `-> str` â†’ function returns a string.

âœ… Python wonâ€™t enforce this at runtime, but type checkers will.

---

## ðŸ”¹ **Q3. Where are annotations stored?**

They are stored in the functionâ€™s **`__annotations__`** dictionary.

```python
print(greet.__annotations__)
# {'name': <class 'str'>, 'age': <class 'int'>, 'return': <class 'str'>}
```

## ðŸ”¹ **Q4. Advanced Annotations with `typing` Module**

```python
from typing import List, Tuple, Dict, Optional

def process(data: List[int], config: Dict[str, str]) -> Tuple[int, int]:
    return min(data), max(data)

def get_name(id: int) -> Optional[str]:
    return "Suraj" if id == 1 else None
```

* `List[int]` â†’ list of integers
* `Dict[str, str]` â†’ dictionary with string keys & values
* `Tuple[int, int]` â†’ tuple containing two ints
* `Optional[str]` â†’ return can be `str` or `None`
