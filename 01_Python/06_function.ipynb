## üîπ **Q1. What is a function in Python?**

**Answer (Interview Style):**
A function is a reusable block of code that performs a specific task. Functions in Python are defined using the `def` keyword and can take inputs (parameters) and return outputs.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Suraj"))  # Hello, Suraj!
```

## üîπ **Q2. Types of Functions in Python**

1. **Built-in Functions** ‚Üí `len()`, `print()`, `sum()`, etc.
2. **User-defined Functions** ‚Üí defined using `def`.
3. **Anonymous Functions (Lambda)** ‚Üí defined using `lambda`.
4. **Recursive Functions** ‚Üí function calling itself.

## üîπ **Q3. Function Arguments in Python**

Python supports **multiple argument types**:

| Type                              | Example                      | Usage                               |
| --------------------------------- | ---------------------------- | ----------------------------------- |
| **Positional**                    | `def add(a, b)` ‚Üí `add(2,3)` | Must match order                    |
| **Keyword**                       | `add(b=3, a=2)`              | Order doesn‚Äôt matter                |
| **Default**                       | `def greet(name="Guest")`    | Uses default if not passed          |
| **Variable Positional (\*args)**  | `def fun(*args)`             | Collects multiple args into a tuple |
| **Variable Keyword (\*\*kwargs)** | `def fun(**kwargs)`          | Collects keyword args into dict     |

```python
def demo(a, b=5, *args, **kwargs):
    print(a, b, args, kwargs)

demo(1, 2, 3, 4, x=10, y=20)
# Output: 1 2 (3,4) {'x':10,'y':20}
```

## üîπ **Q4. What is Scope in Python? (LEGB Rule)**

Python resolves variables using **LEGB**:

* **L** ‚Üí Local (inside function)
* **E** ‚Üí Enclosing (outer functions)
* **G** ‚Üí Global (module-level)
* **B** ‚Üí Built-in (`len`, `print`)

```python
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print(x)
    inner()

outer()  # local
```

## üîπ **Q5. Lambda Functions**

Anonymous, single-expression functions:

```python
square = lambda x: x*x
print(square(5))  # 25
```

Common in `map()`, `filter()`, `sorted()`:

```python
nums = [1,2,3,4]
print(list(map(lambda x: x**2, nums)))  # [1,4,9,16]
```

## üîπ **Q6. Recursion Example**

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(5))  # 120
```

## üîπ **Q7. Higher-Order Functions**

Functions that take other functions as arguments or return them.

```python
def apply_twice(f, x):
    return f(f(x))

print(apply_twice(lambda x: x+1, 5))  # 7
```

---

## üîπ **Q8. Decorators**

Decorators are functions that wrap another function to extend its behavior.

```python
def log(func):
    def wrapper(*args, **kwargs):
        print("Calling:", func.__name__)
        return func(*args, **kwargs)
    return wrapper

@log
def greet(name):
    return f"Hi {name}"

print(greet("Suraj"))
```

## üîπ **Q9. Common Tricky Questions**

1. **Mutable default arguments**

```python
def f(a, lst=[]):
    lst.append(a)
    return lst

print(f(1))  # [1]
print(f(2))  # [1,2]  ‚ö†Ô∏è Same list reused!
```

üëâ Best practice: use `None` as default.

---

2. **Function as First-Class Citizen**

* Functions can be assigned to variables, passed as arguments, and returned.

3. **Difference between `return` and `yield`**

* `return` exits function with a value.
* `yield` turns function into a generator.
