# üîπ Inheritance in Python

**Definition:**
Inheritance allows a class (**child/subclass**) to acquire the properties and behaviors (attributes & methods) of another class (**parent/superclass**).
It promotes **code reuse, modularity, and extensibility**.

---

## ‚úÖ Types of Inheritance in Python

1. **Single Inheritance** ‚Äì child inherits from one parent.
2. **Multiple Inheritance** ‚Äì child inherits from multiple parents.
3. **Multilevel Inheritance** ‚Äì grandchild inherits from child which inherits from parent.
4. **Hierarchical Inheritance** ‚Äì multiple children inherit from the same parent.
5. **Hybrid Inheritance** ‚Äì combination of above (Python uses **MRO: Method Resolution Order** to resolve conflicts).

---

## Example 1: Single Inheritance

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

dog = Dog()
dog.speak()   # Dog barks
```

---

## Example 2: Multiple Inheritance

```python
class Father:
    def skills(self): print("Gardening")

class Mother:
    def skills(self): print("Cooking")

class Child(Father, Mother):
    def skills(self): print("Sports")

c = Child()
c.skills()   # Sports (child overrides)
```

üëâ Python resolves using **MRO (left-to-right search)**.

---

## Example 3: Multilevel Inheritance

```python
class A:
    def show(self): print("Class A")

class B(A):
    def show(self): print("Class B")

class C(B):
    pass

c = C()
c.show()  # Class B
```

# üîπ Interview Q & A on Inheritance

### **Q1. What is inheritance in Python?**

üëâ Inheritance allows a class to reuse attributes and methods of another class, enabling **code reuse and modular design**.

---

### **Q2. What are the types of inheritance in Python?**

üëâ Single, Multiple, Multilevel, Hierarchical, and Hybrid.

---

### **Q3. How does Python handle method conflicts in multiple inheritance?**

üëâ Python uses **MRO (Method Resolution Order)**, which follows the **C3 linearization algorithm**.
You can check it with:

```python
print(Child.__mro__)
```

---

### **Q4. How is inheritance used in ML/AI?**

üëâ

* ML frameworks use inheritance for model design.
* Example: In scikit-learn, all estimators inherit from `BaseEstimator`.
* This ensures consistency across models (`fit()`, `predict()`, `score()`).

---

### **Q5. Difference between inheritance and composition?**

* **Inheritance:** "is-a" relationship (Dog is an Animal).
* **Composition:** "has-a" relationship (Car has an Engine).
  üëâ Composition is often preferred in ML pipelines for flexibility.

---

### **Q6. Can you prevent a class from being inherited?**

üëâ Yes, by raising `TypeError` in `__init_subclass__`.

```python
class Final:
    def __init_subclass__(cls, **kwargs):
        raise TypeError("Subclassing not allowed")
```
