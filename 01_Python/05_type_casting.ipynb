## üîπ **Q1. What is Type Casting in Python?**

**Answer (Interview Style):**
Type casting is the process of converting one data type into another.
Python supports two kinds:

1. **Implicit Type Casting (Type Conversion)** ‚Äì Python automatically converts smaller data types into larger data types to avoid data loss.

   ```python
   x = 5        # int
   y = 2.5      # float
   z = x + y    # Python converts int ‚Üí float
   print(z)     # 7.5 (float)
   ```

2. **Explicit Type Casting (Type Casting)** ‚Äì The programmer manually converts one type to another using built-in functions.

   ```python
   a = "123"
   b = int(a)   # str ‚Üí int
   print(b + 1) # 124
   ```

## üîπ **Q2. What are the common casting functions in Python?**

| Function   | Converts To    | Example                                   |
| ---------- | -------------- | ----------------------------------------- |
| `int(x)`   | Integer        | `int(3.7) ‚Üí 3`                            |
| `float(x)` | Floating point | `float("5") ‚Üí 5.0`                        |
| `str(x)`   | String         | `str(25) ‚Üí "25"`                          |
| `list(x)`  | List           | `list("abc") ‚Üí ['a','b','c']`             |
| `tuple(x)` | Tuple          | `tuple([1,2]) ‚Üí (1,2)`                    |
| `set(x)`   | Set            | `set([1,2,2]) ‚Üí {1,2}`                    |
| `dict(x)`  | Dictionary     | `dict([(1,'a'),(2,'b')]) ‚Üí {1:'a',2:'b'}` |
| `bool(x)`  | Boolean        | `bool([]) ‚Üí False`                        |

## üîπ **Q3. What are some tricky cases with type casting?**

1. **String to int/float requires numeric strings**

   ```python
   int("10")   # ‚úÖ 10
   int("10a")  # ‚ùå ValueError
   ```

2. **Float to int truncates (not rounds)**

   ```python
   int(9.9)   # 9
   ```

3. **Bool conversion follows truthy/falsy rules**

   ```python
   bool(0)     # False
   bool("")    # False
   bool(" ")   # True (non-empty string)
   ```

4. **Casting to dict needs proper format**

   ```python
   dict([(1,"a"), (2,"b")])   # ‚úÖ {1:'a',2:'b'}
   dict([1,2,3])              # ‚ùå TypeError
   ```

## üîπ **Q4. What‚Äôs the difference between `int()` vs `math.floor()` vs `round()`?**

* `int(x)` ‚Üí Truncates towards 0
* `math.floor(x)` ‚Üí Rounds down to nearest integer
* `round(x)` ‚Üí Rounds to nearest integer (ties to even in Python 3)

```python
import math
print(int(-3.7))    # -3
print(math.floor(-3.7))  # -4
print(round(2.5))   # 2 (banker‚Äôs rounding)
```

---

## üîπ **Q5. Why is implicit type casting important in Python?**

Because Python is **dynamically typed**, operations between different types are frequent. Implicit casting prevents data loss and reduces programmer effort. For example, adding `int + float` results in a float, not an error.

## üîπ **Cheat Sheet: Type Casting in Python**

```python
# Implicit Casting
x = 5
y = 2.5
print(x + y)   # 7.5 (int ‚Üí float)

# Explicit Casting
print(int("10"))     # 10
print(float("3.14")) # 3.14
print(str(100))      # "100"

# Collection Conversions
print(list("abc"))   # ['a','b','c']
print(tuple([1,2]))  # (1,2)
print(set([1,2,2]))  # {1,2}

# Boolean Conversion
print(bool(0))       # False
print(bool("Hi"))    # True

# Dict Conversion
pairs = [(1,"a"),(2,"b")]
print(dict(pairs))   # {1:'a',2:'b'}
```
